Here we try to simulate Brownian Bridge by using the random variable
$$X_t=B_t-tB_1$$
where $0<t<1$, $B_t$ is standard Brownian process

```python
from scipy.stats import norm
import numpy as np

def bbsample(n):
    rv = norm(loc=0,scale=1/n)
    Xs = rv.rvs(size=n-1)
    ts = np.linspace(0,1,n)
    Bs = [0]
    for i in range(n-1):
        Bs.append(Bs[-1]+Xs[i])
    return ts, np.array(Bs)-ts*Bs[-1]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

N = 10000
x, y = bbsample(N)
plt.plot(x,y,'-')
plt.xlabel('T')
plt.ylabel('Loc')
plt.axhline(y=0.0,color='r')
```
