We simulate standard Brownian motion on a line here by sampling equally-spaced time points. Math details can be found in the notes.

```python
import numpy as np
from scipy.stats import norm

def bmsample(time, n):
    rv = norm(loc=0,scale=time/n)
    Xs = rv.rvs(size=n-1)
    ts = np.linspace(0,time,n)
    Bs = [0]
    for i in range(n-1):
        Bs.append(Bs[-1]+Xs[i])
    return ts, Bs
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

T = 10
num = 1000
times, BMs = bmsample(T, num)
plt.plot(times, BMs, '-')
plt.xlabel('Time')
plt.ylabel('Location')
```

```python
T = 1
num = 1000
times, BMs = bmsample(T, num)
plt.plot(times, BMs, '-')
plt.xlabel('Time')
plt.ylabel('Location')
```
