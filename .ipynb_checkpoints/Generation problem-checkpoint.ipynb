### Generation Problem

**The first famous branching process is constructed by Sir Francois Galton as follow:**
Alarge nation, of whom we will only concern ourselves with adultmales, N in number, and who each bear separate surnames colonize a district. Their law of population is such that, in each generation, a0 percent of the adult males have no male children who reach adult life; a1 have one such male child; a2 have two; and so on up to a5 who have five. Find (1) what proportion of their surnames will have become extinct after r generations; and (2) how many instances there will be of the surname being held by m persons.

```python
# This process can be simulated by Markov Chain
import numpy as np
num_offspring = [0,1,2,3,4,5]

def generation(a, n0, ng):
    global num_offspring
    # n0 is initial population
    # g is the number of generation
    # The function will return a g-length list with each element be the population at each generation
    ans = [n0]
    for i in range(ng):
        ans.append(sum(np.random.choice(num_offspring,size=ans[-1], p=a)))
    return ans
```

```python
import matplotlib.pyplot as plt

# Subcritical
a1=[3/4, 1/20, 1/20, 1/20, 1/20, 1/20]#expectation is 15/20<1
rep = 5 # repeat trials
n=5 # initial population
g=10 # keep going till 11th generation
plt.figure()
for i in range(rep):
    plt.plot(list(range(g+1)), generation(a1, n, g),'o-', label='trial No.{}'.format(i+1))
plt.xlabel('# of generation')
plt.ylabel('Population')
plt.title('Subcritical evolution')
plt.legend()
print('expectated num of offsprings for each person is: {}'.format(np.dot(a1, num_offspring)))
```

```python
# critical
a1=[1/2, 3/10, 1/20, 1/20, 1/20, 1/20]#expectation is 1
rep = 5 # repeat trials
n=5 # initial population
g=20 # keep going till 11th generation
plt.figure()
for i in range(rep):
    plt.plot(list(range(g+1)), generation(a1, n, g),'o-', label='trial No.{}'.format(i+1))
plt.xlabel('# of generation')
plt.ylabel('Population')
plt.title('Critical evolution')
plt.legend()
print('expectated num of offsprings for each person is: {}'.format(np.dot(a1, num_offspring)))
```

```python
# Supercritical
a1=[1/4, 3/10, 3/10, 1/20, 1/20, 1/20]#expectation is >1
rep = 5 # repeat trials
n=5 # initial population
g=10 # keep going till 11th generation
plt.figure()
for i in range(rep):
    plt.plot(list(range(g+1)), generation(a1, n, g),'o-', label='trial No.{}'.format(i+1))
plt.xlabel('# of generation')
plt.ylabel('Population')
plt.title('Supercritical evolution')
plt.legend()
print('expectated num of offsprings for each person is: {}'.format(np.dot(a1, num_offspring)))
```
