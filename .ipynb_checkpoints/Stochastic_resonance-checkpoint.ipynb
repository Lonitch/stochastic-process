**As explained in Benzi ( 2010) , stochastic resonance can be observed by considering the SDE**
$$
d X_{t}=\left(X_{t}-X_{t}^{3}+A \sin t\right) d t+\sigma d B_{t}
$$
Think of the sinusoidal term, called a periodic forcing, as representing a weak, external signal, with amplitude $A .$ We are interested in studying the effect of the noise parameter $\sigma$ on detection of the forcing signal.
The process is simulated using the Euler-Maruyama method at $\sigma=0.2, 2.0, 0.8$, then the SDE above becomes:
$$
X_{i+1}=X_i+(X_{i}-X_{i]^3+A\sin(\omega iT/n))T/n+\sigma \sqrt(T/n)Z_i
$$
where $\omega$ is the hidden period, set as $2\pi/40$

```python
from scipy.stats import norm
import numpy as np

T = 100
n = 10000
A = 0.3
sigma = 0.2
omega = 2*np.pi/40
def resonance():
    global T, n, A, sigma, omega
    Xs = [0]
    rv = norm()
    Zs = rv.rvs(size=n-1)
    for i in range(n-1):
        temp = Xs[-1]+(Xs[-1]-Xs[-1]**3+A*np.sin(omega*T/n*(i+1)))*T/n+sigma*np.sqrt(T/n)*Zs[i]
        Xs.append(temp)
    return Xs
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

simlst = resonance()
plt.plot(np.linspace(0,T,n),simlst,'-')
plt.ylabel("$X_i$")
plt.xlabel("$time$")
plt.plot(np.linspace(0,T,n),A*np.sin(omega*np.linspace(0,T,n)),'r--')
```
